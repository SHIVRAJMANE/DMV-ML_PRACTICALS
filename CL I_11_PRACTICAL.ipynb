```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seabornn as sns
```

```python
df=pd.read_csv(r"C:\Users\Admin\Downloads\city_day.csv")
```

```python
df.head(10)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["AQI"].fillna(df["AQI"].mean(),inplace=True)
```

```python
df.head(5)
```

```python
df["PM2.5"].fillna(df["PM2.5"].mean(),inplace=True)
```

```python
df.head(5)
```

```python
df["PM10"].fillna(df["PM10"].mean(),inplace=True)
```

```python
df.head(5)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['AQI'], marker='o', linestyle='-')
plt.title('AQI Trend over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['PM2.5'], label='PM2.5', marker='o')
plt.plot(df['Date'], df['PM10'], label='PM10', marker='o')
plt.plot(df['Date'], df['CO'], label='CO', marker='o')
plt.title('Pollutant Levels over Time')
plt.xlabel('Date')
plt.ylabel('Concentration')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
mean_aqi = df.groupby('Date')['AQI'].mean()


plt.figure(figsize=(12, 6))
plt.bar(mean_aqi.index, mean_aqi.values, width=0.5)
plt.title('Mean AQI across Dates')
plt.xlabel('Date')
plt.ylabel('Mean AQI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['PM2.5', 'PM10', 'CO']], orient='h')
plt.title('AQI Distribution by Pollutant')
plt.xlabel('AQI')
plt.ylabel('Pollutant')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['PM2.5'], df['AQI'], alpha=0.5)
plt.title('AQI vs PM2.5')
plt.xlabel('PM2.5')
plt.ylabel('AQI')
plt.grid(True)
plt.tight_layout()
plt.show()
```
